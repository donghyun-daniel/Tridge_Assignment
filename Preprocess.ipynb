```python
import pandas as pd
import re
```

```python
data = pd.read_csv("Dataset_-_Dataset.csv")
data.head()
```

Product col의 요소 분리, Variety, Grades에서의 필터링 및 추출, Region 정리, 날짜의 정렬이 필요해보인다

먼저 Product col을 Preprocessing 진행

```python
data["Product"] = data["Product"].str.findall(r'^(\D+) \(#(\d+)\)$')
data.head()
```

```python
data["Product_Name"] = data["Product"].apply(lambda x : x[0][0])
data["Product_Number"] = data["Product"].apply(lambda x : x[0][1])
data.drop(["Product"], axis = 1, inplace = True)
data.head()
```

의도대로 Product_Name, Product_Number가 신규 생성되었다

Variety와 Grade의 경우, 가장 마지막에 기재된 항목만을 데이터로 가져야 종속성을 해치지 않으면서 atomic을 가질 수 있을 것으로 보인다

```python
data["Variety"] = data.Variety.str.split("/ ").str[-1]
data["Grades"] = data.Grades.str.split("/ ").str[-1]
data.head()
```

Region은 일관성이 없게 수집된 것을 볼 수 있다. ,를 delimeter로 보고, (A, B) 구성일 경우 B가 더 큰 개념인 것을 볼 수 있는데, B가 존재하지 않는 경우도 있다. '

일관성을 위해 (A, B) 구성으로 사용할 지, Dallas, Texas의 경우 Texas만 사용할 지 등에 대해 일관된 Rule이 필요해 보인다.
만약 (A, B) 구성이 반드시 필요하다면, 소분류 지역과 대분류 지역을 mapping한 Table을 활용해 (A, B)와 같이 처리해 줄 수 있을 것으로 보인다
