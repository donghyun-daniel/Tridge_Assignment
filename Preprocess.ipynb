```python
import pandas as pd
import re
```

```python
data = pd.read_csv("Dataset_-_Dataset.csv")
data.head(500)
```

```python
for idx, row in data.iterrows():
    [(name, num)] = re.findall(r'^(\D+) \(#(\d+)\)$', data["Product"][idx])
    data["Product_name"][idx] = name
    data["Product_number"][idx] = num
    #pd.DataFrame(data["Product"].str.findall(r'^(\D+) \(#(\d+)\)$', 'Avocado (#99)'))
```

Variety와 Grade의 경우, 가장 마지막에 기재된 항목만을 데이터로 가져야 종속성을 해치지 않으면서 atomic을 가질 수 있을 것으로 보인다

```python
data["Variety"] = data.Variety.str.split("/ ").str[-1]
data["Grades"] = data.Grades.str.split("/ ").str[-1]
```
